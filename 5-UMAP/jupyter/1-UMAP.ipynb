```python
import pandas as pd
from umap.umap_ import UMAP
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from aquarel import load_theme
theme = load_theme('boxy_light')

# 忽略所有来自 sklearn.deprecation 的 FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)
# 忽略所有 UMAP 的 UserWarning
warnings.filterwarnings("ignore", category=UserWarning)

# 1. 读取数据
df = pd.read_csv(f'{BASE_DIR}/hpglobal_LD_PCA.eigenvec.hspAEAsia_UMAP.csv')

# 2. 提取 PC1–PC10（假设在第3–12列）
X = df.iloc[:, 2:12].values

# 3. 运行 UMAP
reducer = UMAP(
    n_neighbors=10, # 设置邻居数量
    min_dist=0.1, # 设置最小距离
    n_components=2, # 降维到2维
    metric='euclidean', # 使用欧氏距离
    init='spectral', # 使用谱初始化
    random_state=42 # 设置随机种子以确保结果可重复
)
embedding = reducer.fit_transform(X)

# 4. 构建可绘图 DataFrame
id_col = df.columns[0]
out_df = pd.DataFrame({
    id_col:       df[id_col],
    'V1':         embedding[:, 0],
    'V2':         embedding[:, 1],
    'group':      df['group'].astype('category')
})

# 5. 使用高区分度调色板（如tab20、Set3等）避免颜色重复
unique_groups = out_df['group'].cat.categories
num_groups = len(unique_groups)
# tab20最多支持20种颜色，Set3支持12种，若类别更多可自定义
if num_groups <= 20:
    palette = sns.color_palette('tab20', num_groups)
else:
    palette = sns.color_palette('hls', num_groups)
palette_dict = dict(zip(unique_groups, palette))

# 6. 使用 seaborn 绘图
theme.apply()
plt.rcParams['font.family'] = ['Arial'] # 设置字体
plt.rcParams['pdf.fonttype'] = 42 # 设置PDF字体类型
plt.rcParams['ps.fonttype'] = 42 # 设置PS字体类型
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=out_df,
    x='V1', y='V2',
    hue='group',
    palette=palette_dict,
    s=50, # 点的大小
    alpha=0.7, # 点的透明度  
    edgecolor='#FFFFFF', # 点的边框颜色
    linewidth=0, # 描边粗细为 0.5
    legend='full'
)
plt.grid(False) # 关闭网格线
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.title('UMAP')
plt.legend(title='')
plt.tight_layout()
plt.show()
```

```python
# 手动剔除一些离群值
out_df_no_outlay = out_df.loc[(out_df['group'] == 'hspHAEAsiaS') & (out_df['V1'] < 0) |
           (out_df['group'] == 'hspLAEAChinaNS') & (out_df['V1'] > -1) & (out_df['V2'] < 10) & (out_df['V1'] < 10)|
           (out_df['group'] == 'hspLAEAJapan') & (out_df['V1'] > 0) |
           (out_df['group'] == 'hspLASEAsia') & (out_df['V1'] > 5)  & (out_df['V2'] > 5)   |
           (out_df['group'] == 'hspHAEAsiaN') & (out_df['V1'] < 0 ) |
           (out_df['group'] == 'hspLAEAChinaN') |
           (out_df['group'] == 'hspLAEAChinaS') & (out_df['V1'] > -5) & (out_df['V2'] < 5)
           ]
```
